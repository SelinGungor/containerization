# Docker Volumes

Volumes are the preferred mechanism for persisting data generated by and used by Docker containers and to mount data from other sources into the Docker containers.



Where to use VOLUME?
- databases
- web servers
- yarn data nodes



#### Create a volume:

`docker volume create my-vol`

#### List volumes:

`docker volume ls`


#### Inspect a volume:

`docker volume inspect my-vol`
